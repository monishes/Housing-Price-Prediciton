### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing the Datasets

```python
dataset=pd.read_csv("HousingPrices.csv")
X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
y=y.reshape(len(y),1)
print(X)
print(y)
```

### Deriving our Polynomial features with degree 2

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
X_poly=poly.fit_transform(X)
```

### Splitting the Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_poly,y,test_size=0.25,random_state=0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
sc_y=StandardScaler()
X_train=sc_x.fit_transform(X_train)
X_test=sc_x.transform(X_test)
y_train=sc_y.fit_transform(y_train)
y_test=sc_y.transform(y_test)
```

### Training our Polynomial Regression Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

### Predicting the Test set results

```python
y_pred=sc_y.inverse_transform(regressor.predict(X_test))
```

### Checking the R-Squared value for our model

```python
from sklearn.metrics import r2_score
print(r2_score(sc_y.inverse_transform(y_test),y_pred))
```

#### Our Polynomial Regressionmodel has an Accuracy of 84.28%
